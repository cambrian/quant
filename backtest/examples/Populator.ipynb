```python
data_dir = 'data'
# Second item in tuple is max batch size for requests.
exchanges = [('binance', 1000), ('bitmex', 200)]
pairs = ['BTC/USDT', 'ETH/USDT', 'BTC/USD', 'ETH/USD']
tick_size = '1m'
start = '2019-03-01T00:00:00Z'
num_ticks = 12 * 60
```

```python
populator.populate(data_dir, exchanges, pairs, tick_size, start, num_ticks)
```

```python
btc_usdt_frame = populator.load_data_as_frame(data_dir, 'binance', 'BTC/USDT', tick_size, start, num_ticks)
btc_usdt_frame
```

```python
import backtrader as bt

class SimpleRsiStrategy(bt.Strategy):
    def __init__(self):
        self.rsi = bt.indicators.RSI_SMA(self.data.close, period=21)

    def next(self):
        if not self.position:
            if self.rsi < 30:
                self.buy(size=10)
        else:
            if self.rsi > 70:
                self.sell(size=10)

cerebro = bt.Cerebro()
cerebro.addstrategy(SimpleRsiStrategy)
data = bt.feeds.PandasData(dataname=btc_usdt_frame)
cerebro.adddata(data)

starting_cash = 100000
cerebro.broker.setcash(starting_cash)
cerebro.run()

# End diagnostics.
portfolio_value = cerebro.broker.getvalue()
pnl = portfolio_value - starting_cash

print('Final Portfolio Value: ${}'.format(portfolio_value))
print('P/L: ${}'.format(pnl))

cerebro.plot(style='candlestick')
```
